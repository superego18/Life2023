```python
from datetime import datetime

def record_time():
    now = datetime.now()
    formatted_date = now.strftime("%Y%m%d%H")
    return formatted_date
```

```python
import os 

def download_stockDataset(con, tickers, marketName):
    folder_name = f"downloads-{marketName}-{record_time()}"
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)
    bd_columns = ['PX_LAST', 'CHG_PCT_1D', 'VOLUME']
    start_date = '20150101'
    end_date = '20230706'
    for ticker in tickers:
        df = con.bdh(f'{ticker} Equity', bd_columns, start_date, end_date)
        columns = [column for (ticker,column) in df.columns]
        df.columns = columns
        df.to_csv(f'./{folder_name}/dataset-{makretName}-{ticker}.csv')
```
